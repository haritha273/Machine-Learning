```python
#Encoding 
#drawback of label encoding==form a hierrarcy 
#apple =0 orange =1 ,mango=2 ,grape=3,mango=2,grape=3,mango=2
#  apple has highest price in market but in encoding apple has lower value ,ie form a hierrarchy so label encoding has this drawback 
#to avoid this use 0ne -Hot Encoding==convert all in to different columns 
```

```python
from sklearn.compose import make_column_transformer  #to convert into different columns
from sklearn.preprocessing import OneHotEncoder
col_trans=make_column_transformer((OneHotEncoder(handle_unknown='ignore'),['State']),remainder='passthrough')
x=col_trans.fit_transform(x)
x
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred


```

```python
y_test
```

```python
df2=pd.DataFrame({'Actual_value':y_test,'predicted_value':y_pred,'Difference':y_test-y_pred})
df2
```

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score
MAPE=mean_absolute_percentage_error(y_test,y_pred)
MAPE
```

```python
score=r2_score(y_test,y_pred)
score
```
