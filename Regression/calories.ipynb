```python
outlier_columns=['Height','Weight','Heart_Rate','Body_Temp','Calories']
def remove_outliers(df,columns):
    for i in columns:
        Q1=df[i].quantile(0.25)
        Q3=df[i].quantile(0.75)
        IQR=Q3-Q1
        lower_bound=Q1-1.5*IQR
        upper_bound=Q3+1.5*IQR
        df=df[(df[i]>=lower_bound)&(df[i]<=upper_bound)]
    return df
df_cleaned=remove_outliers(df,outlier_columns)
```

```python
correlation_matrix=df_cleaned.corr()
plt.figure(figsize=(12,8))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',fmt='.2f',linewidths=0.5) #annot=true,to write value in cells,bydefautlt it is falt
plt.title('Heatmap')
```

```python
df_selected=df_cleaned.drop(columns=['Height','Gender'])
correlation_matrix=df_selected.corr()
plt.figure(figsize=(12,8))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',fmt='.2f',linewidths=0.5)
plt.title('Heatmap')
```

```python
x=df_selected.iloc[:,:-1]
x
```

```python
y=df_selected.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score
MAPE=mean_absolute_percentage_error(y_test,y_pred)
MAPE
```
