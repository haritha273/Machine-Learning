```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('company_sales_data.csv')
df
```

```python
# 1. Read Total profit of all months and show it using a line plot
x=df['month_number']
y=df['total_profit']
plt.plot(x,y)
plt.xlabel('Month NUmber')
plt.ylabel('Total profit')
plt.xticks(x)
plt.yticks([100000,200000,300000,400000,500000,600000,700000,800000])
plt.legend(loc='lower right')                   #define values of x axis and y axis
```

```python
# 2. Get total profit of all months and show line plot with the following Style properties
#  Generated line plot must include following Style properties:
#  Line Style dotted and Line-color should be red
#  Show legend at the lower right location.
#  X label name = Month Number
#  Y label name = Sold units number
#  Add a circle marker.
#  Line marker color as read
#  Line width should be 3
x=df['month_number']
y=df['total_profit']
plt.plot(x,y,linestyle='dotted',color='r',marker='o',mfc='r',mec='r',linewidth=3,label=' Profit Data')
plt.xlabel('Month NUmber')
plt.ylabel('Total profit')
plt.xticks(x)
plt.yticks([100000,200000,300000,400000,500000])
plt.legend(loc='lower right')
```

```python
#3. Read all product sales data and show it  using a multiline plot
#     square marker with color=green 
facecream=df['facecream']
facewash=df['facewash']
tootpaste=df['toothpaste']
bath=df['bathingsoap']
sham=df['shampoo']
moist=df['moisturizer']
plt.plot(x,facecream,marker='s',color='g',label='Face cream sales')
plt.plot(x,facewash,marker='s',color='k',label='Fcewash sales')
plt.plot(x,tootpaste,marker='s',color='c',label='Tooth Paste sales')
plt.plot(x,bath,marker='s',color='r',label='Bathingsoap sales')
plt.plot(x,sham,marker='s',color='b',label='Shampoo sales')
plt.plot(x,moist,marker='s',color='g',label='Moisturizer sales')
plt.xlabel('Month Number')
plt.ylabel('Total sales of each category')
plt.legend(loc='upper right')
plt.xticks(x)
plt.yticks([1000,2000,3000,4000,5000,6000,7000,8000,9000,10000,11000,12000,13000,14000])
```

```python

# 4.Read toothpaste sales data of each month and show it using a scatterplot
plt.scatter(x,tootpaste)
```

```python
# 5.Read face cream and facewash product sales data and show it using the bar chart with grid line
plt.bar([a-0.25 for a in x],facecream,width=0.25,align='edge',label='Face Cream Sales')
plt.bar([a+0.25 for a in x],facewash,width=0.25,align='edge',label='Face Cream Sales')
plt.xlabel('Month Number')
plt.ylabel('Profit')
plt.title('Face Cream and Face Wash')
plt.legend(loc='upper right')
plt.tight_layout()
plt.xticks(x)
plt.grid()
```

```python
labels=['Face Cream','Face Wash','Bathing Soap','Tooth Paste','Shamboo','Moisturizer']
sales=[df['facecream'].sum(),df['facewash'].sum(),df['bathingsoap'].sum(),df['tootpaste'].sum(),df['Shamboo'].sum(),df['Moisturizer'].sum()]
plt.pie(sales,labels=labels,autopct='%1.01f%%')
plt.legend(loc='upper right')
plt.tight_layout()
```
