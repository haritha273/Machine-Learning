```python
import numpy as np
import pandas as pd
df=pd.read_csv('customer1.txt',header=None)
df.columns=['id','fname','lname','age','prof','loc']
df
```

```python
df1=df['loc'].value_counts()
df1
```

```python
#collect columns from a dataframe ==df1[],pass index and values of corresponding columns
plt.bar(df1.index,df1.values)  #(generate values corre to index)
```

```python
plt.bar(df1.index,df1.values)
plt.xlabel('Locations',color='r')
plt.ylabel('Count',color='r')
plt.title('Location Count Data',color='k')
```

```python
plt.pie(df1.values,labels=df1.index,autopct='%1.0f%%')
plt.legend(loc='upper right')
plt.tight_layout()
```
