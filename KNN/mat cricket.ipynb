```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
runs=[16,10,12,8,10,12,10,6,8,18,20,2,8,8,14,10,8,10,12,16]
wickets=[0,0,1,0,0,0,0,0,2,0,0,0,3,0,0,0,0,0,0,1]
plt.bar(range(1,len(runs)+1),runs,label='Runs',color='g')
wicket_count = 0  
for i in range(len(wickets)):
  if wickets[i]>0:
    for j in range(1,wickets[i]+1):
      wicket_count+=1
      if wicket_count!=sum(wickets):
        plt.scatter(i+1,runs[i]+j, color='red')
      else:
        plt.scatter(i+1,runs[i]+j, color='red',label="Wickets")
        plt.xlabel('Overs')
plt.ylabel('Runs per over')
plt.xticks(range(1,len(runs)+1))
plt.yticks(range(0,max(runs)+6,4))
plt.title('Score Analysis', fontsize=15, fontweight='bold', fontstyle='italic')
plt.legend()
plt.show
```
