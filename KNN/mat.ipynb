```python
#to mark a points without draw a line 
x=[5,2,3,4]
y=[8,9,10,11]
plt.plot(x,y,'*') #* act as a mark can give any value +,*,o=point(),s=(square)
```

```python
x1=np.array([1,2,3,4,5,6])
y1=np.array([3,8,1,10,5,7]) 
plt.plot(x1,y1,marker='o') #it give line as well as the points 
```

```python
plt.plot(x1,y1,'*:r')  #*:r==* marker,:-line (dotted line),r-colur r represent red color 
#3 types of line ==solid line,2 dotted line ,3,dashed line 

#1.marker ==*,+,o,s
#2.:line :(dotted),--(dotted)
#3.colour r-red ,m -magenta,b-blue ,g-green ,y -yellow ,k-black,c-cyan
```

```python
#ms=marker size ,we can change size of marker size using ms
#mfc=marker face colour (marker has face and edge)
#mec=marker edge colour 
plt.plot(x1,y1,marker='o',ms=21,mfc='r',mec='m')
```

```python
#properties of line 
#1.linestyle 1.solid -if we not enter the linestyle machine itself give solid line
              #2dotted ,
              # 3 ,dashed
#2 width of line =linewidth (can chage 10,8,7 ect )
#3 colour - can change colour of line using color 
plt.plot(x1,y1,marker='o',linestyle='dotted',linewidth=10,color='g')
```

```python
#content in the x and y axis=plt.xlabel,plt.ylabel
height=[150,156,157,158]
weight=[45,56,47,57]
plt.plot(height,weight)
plt.xlabel('height in cm',color='r',fontsize=10)
plt.ylabel('weight in kg',color='r',fontsize=10)
plt.title('height-weight graph',fontsize=15)
#plt.grid()
#plt.grid(axis='x')# line per to x axis
plt.grid(axis='y')#line per to y axis
```

```python
#velocity dis graph
Velocity=[4,5,6,7]
Displacement=[12,13,14,15]
plt.plot(Velocity,Displacement)
plt.xlabel('Velocity in m/s',color='g',fontsize=10)
plt.ylabel('Displacement in m',color='g',fontsize=10)
plt.title('Velocity-Displacement Graph',fontsize=20)
plt.grid(linestyle='dotted',linewidth=3,color='r') #to convert to grid from

```

```python
#scatter plot =to represent each point 
x1=[1,4,6,7,8,9,10]
y1=[4,6,2,5,3,2,5]
x2=[1,5,7,6,5,4,3]
y2=[3,4,5,6,7,8,2]
plt.scatter(x1,y1,color='r',marker='*')
plt.scatter(x2,y2,color='g',marker='s')
```

```python
#subplot--convert a graph in to two one for graph 1 and another for graph 2 
x1=[1,2,3,4]
y1=[3,7,1,9]
plt.subplot(1,2,1)#there  is 3 argument =3rd argument(1)=position of graph ,if give 1 ,first graph appered in first position ,(1,2)=there is 1 row and 2 colmns
plt.plot(x1,y1) #if give 1 st and sec arguments as (2,1) get 2 rows and 1 column
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.title('Graph1')


x2=[3,4,6,1]
y2=[1,7,2,8]
plt.subplot(1,2,2)#3rd arg=2,second graph second position ,if change that position graph also change
plt.plot(x2,y2)
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.title('Graph2')
#supertitle==give title for both graph
plt.suptitle('Graph1-Graph2')
```
