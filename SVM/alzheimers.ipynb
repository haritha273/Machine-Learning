```python
#encoding
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
df['Country']=lab.fit_transform(df['Country'])
df['Gender']=lab.fit_transform(df['Gender'])
df['Physical Activity Level']=lab.fit_transform(df['Physical Activity Level'])
df['Smoking Status']=lab.fit_transform(df['Smoking Status'])
df['Alcohol Consumption']=lab.fit_transform(df['Alcohol Consumption'])
df['Diabetes']=lab.fit_transform(df['Diabetes'])
df['Hypertension']=lab.fit_transform(df['Hypertension'])
df['Cholesterol Level']=lab.fit_transform(df['Cholesterol Level'])
df['Family History of Alzheimer’s']=lab.fit_transform(df['Family History of Alzheimer’s'])
df['Depression Level']=lab.fit_transform(df['Depression Level'])
df['Sleep Quality']=lab.fit_transform(df['Sleep Quality'])
df['Dietary Habits']=lab.fit_transform(df['Dietary Habits'])
df['Air Pollution Exposure']=lab.fit_transform(df['Air Pollution Exposure'])
df['Employment Status']=lab.fit_transform(df['Employment Status'])
df['Marital Status']=lab.fit_transform(df['Marital Status'])
df['Genetic Risk Factor (APOE-ε4 allele)']=lab.fit_transform(df['Genetic Risk Factor (APOE-ε4 allele)'])
df['Social Engagement Level']=lab.fit_transform(df['Social Engagement Level'])
df['Income Level']=lab.fit_transform(df['Income Level'])
df['Stress Levels']=lab.fit_transform(df['Stress Levels'])
df['Urban vs Rural Living']=lab.fit_transform(df['Urban vs Rural Living'])
```

```python
df
```

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.30,random_state=0)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import MinMaxScaler
scalar=MinMaxScaler()
scalar.fit(x_train)
x_train=scalar.transform(x_train)
x_test=scalar.transform(x_test)
x_train

```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
score=accuracy_score(y_test,y_pred)
score
```
