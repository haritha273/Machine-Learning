```python
import matplotlib.pyplot as plt
def removal_box_plot(df,column,threshold):
    sns.boxplot(df[column])
    plt.title(f"Original box plot of {column}")
    plt.show()

    removed_outliers=df[df[column]<=threshold]

    sns.boxplot(removed_outliers[column])
    plt.title(f'Box Plot without Outliers of {column}')
    plt.show()
    return removed_outliers

threshold_value=40
no_outliers=removal_box_plot(df,'total_bill',threshold_value)
    
```

```python
import numpy as np
Q1=np.percentile(df['total_bill'],25,method='midpoint')
Q3=np.percentile(df['total_bill'],75,method='midpoint')
IQR=Q3-Q1
print(IQR)

upper=Q3+1.5*IQR
upper_array=np.array(df['total_bill']>=upper)
print('upper bound',upper)
print(upper_array.sum())

#below lower bound
lower=Q1-1.5*IQR
lower_array=np.array(df['total_bill'])<=lower
print('lower bound',lower)
print(lower_array.sum())
```

Outlayers removal methods in box plot

1.manual method
2.Capping or Winsorizing
Instead of removing outliers, you can cap them at the lower and upper bounds.
3. Transformations
Applying a transformation (like logarithmic, square root, or Box-Cox transformation) can reduce the influence of outliers.
4.Use Robust Statistical Methods
Instead of removing or transforming the data, use models that are less sensitive to outliers, such as:
Robust regression models (like Huber or RANSAC in machine learning)
Median instead of mean when aggregating data
5. Imputation
Replace outliers with the mean or median of the data.



Chi-Square (Ï‡Â²) Test
*To find relationship between two categorical columns
*The Chi-Square test evaluates if two categorical variables are independent.
*Seaborn helps visualize relationships, but the actual test is done using scipy.stats.chi2_contingency.
*If the p-value is small (e.g., < 0.05), you reject the null hypothesis and conclude that there is a significant relationship between the two variables.
